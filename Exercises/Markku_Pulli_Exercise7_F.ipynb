```python
##>>> Some code for bootstrap
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

from sklearn import datasets 
from mpl_toolkits.mplot3d import Axes3D

from IPython.core.interactiveshell import InteractiveShell #enables multiple cell outputs withput print command.
InteractiveShell.ast_node_interactivity = "all"
```

### Task 1: MLPRegressor

Implement a `MLPRegressor` to predict `y` using data in `X`. Try different number of hidden layers and number of perceptrons in the hidden layer to reach good prediction accuracy measured by `r2_score` and `cross_val_score`. Avoid too high complexity so that prediction in the training set would be significantly better than in the cross-validation set.   

```python
from sklearn.neural_network import MLPRegressor
from sklearn.metrics import r2_score
from sklearn.model_selection import cross_val_score
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import make_scorer
from sklearn.metrics import accuracy_score

# YOUR CODE HERE
scoring = {"AUC": "roc_auc", "Accuracy": make_scorer(accuracy_score)}
predictor=MLPRegressor(max_iter=1000) #just to pass the assert 
#Use CV to serach best tune parameters.
hidden_layer_sizes=[25,50,100,150,200,250,300,350,400,500,600,1000]
clf = GridSearchCV(MLPRegressor(random_state=1, max_iter=200,verbose=0),
                   param_grid = {'hidden_layer_sizes':hidden_layer_sizes}, 
                   return_train_score=True)
#clf=predictor
%time clf.fit(X,y)
preds =clf.predict(X)
accuracy = clf.score(X,y)
accuracy_cv = cross_val_score(clf,X,y,cv=5).mean()
results = clf.cv_results_
print("Prediction accuracy in training set is", accuracy)
print("Prediction accuracy in cv is          ", accuracy_cv)
#test how long it takes to fit the model
%time predictor.fit(X,y)

#Training time and parameter set
predictor=MLPRegressor(max_iter=1000)
%time predictor.fit(X,y)
accuracy = clf.score(X,y)
accuracy
clf.get_params()

#Visualization
layer_size=[25,50,100,150,200,250,300,350,400,500,600,1000]
fig, ax = plt.subplots()
ax2 = ax.twinx()
ax.plot(layer_size, results['mean_train_score'], label='mean_train_score',linestyle='-', color='red')
ax2.plot(layer_size, results['std_train_score'], label='std_train_score',linestyle='--',color='red')
ax.plot(layer_size, results['mean_test_score'], label='mean_test_score',linestyle='-')
ax2.plot(layer_size, results['std_test_score'], label='std_test_score',linestyle='--')
fig.legend(loc='center left', bbox_to_anchor=(1, 0.2))
plt.title('Accuracy score with different hidden layer sizes')
ax.set_xlabel('Hidden layers')
ax.set_ylabel('mean score')
ax2.set_ylabel('std score')
#ax2.legend()
plt.show()
```
